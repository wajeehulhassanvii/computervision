```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline
```

```python
image = cv2.imread("./sunflower_image_06_hd_pictures_167025.jpg")
image_copy = np.copy(image)
```

```python
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_copy)
```

```python
gray_img = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
```

# canny edge detection implementation

```python
# 1:2 or 1:3 ration between lower and upper for best result 
lower = 120
upper = 240

edges_img = cv2.Canny(gray_img, lower, upper)
plt.imshow(edges_img, cmap='gray')
```

```python
# cutt off for weak is higher in tight and lower in wide
wide_img = cv2.Canny(gray_img, 30, 100)
tight_img = cv2.Canny(gray_img, 180, 240)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.set_title("wide")
ax1.imshow(wide_img, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight_img, cmap='gray')
```
